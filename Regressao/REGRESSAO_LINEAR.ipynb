```python
import numpy as np 
from numpy.random import randn 
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

#### Dados do famoso dataset 'boston houses'

```python
df = pd.DataFrame(load_boston().data, columns = load_boston().feature_names)
df['MEDV'] = load_boston().target
df.head()
```

#### 1.1 - Regressão linear simples - método dos mínimos quadrados 

```python
def reg_lin_simples(x,y):
    x, y = np.array(x), np.array(y)
    x_mean, y_mean = np.mean(x), np.mean(y)
    beta_1 = np.dot(x - x_mean, y - y_mean)/np.linalg.norm(x - x_mean)**2 
    beta_0 = y_mean - beta_1*x_mean
    return [beta_0, beta_1]
```

```python
x = df.AGE.values
y = df.MEDV.values 

beta_0, beta_1 = reg_lin_simples(x,y)
print("beta_0 =", beta_0)
print("beta_1 =", beta_1)

y_pred_mmq = x*beta_1 + beta_0
```

#### 1.2 - Regressão linear simples - usando framework sklearn 

```python
# x deve sempre ser uma matriz ao usar o scikit-learn
X = df.AGE.values.reshape((len(df.AGE), 1)) 
y = df.MEDV 

# Criando e treinando um modelo
modelo = LinearRegression().fit(X, y)
print("beta_0 = ", modelo.intercept_) 
print("beta_1 = ", modelo.coef_[0]) 

y_pred_skl = modelo.predict(x)
```

```python
X = df.iloc[:,:-1]
X = sm.add_constant(X).values 
y = df.MEDV.values 

B = np.dot(np.linalg.pinv(np.dot(X.T, X)), np.dot(X.T,y))

# Imprimindo os coeficientes 
for i in range(len(B)):
    print("beta_%s = %f" %(i, B[i]))
    
y_pred_mmq = np.dot(X,B)
```
